```python
def selection_sort_maior(lista):
    n = len(lista)
    for slot in range(n-1, 0, -1):
        pos_max = 0
        for loc in range(1, slot + 1):
            print('slot analisado ', slot, 'item no slot ', lista[slot], 'loc de comparacao ', loc, 'item de comp ', lista[loc])
            if lista[loc] > lista[pos_max]:
                print(lista[loc], ' é maior que ', lista[pos_max])
                pos_max = loc
            else:
                print(lista[loc], ' não é maior que ', lista[pos_max])
        print('maior item de lá pra cá:', lista[pos_max])
        print('antes ', lista)
        temp = lista[slot]
        lista[slot] = lista[pos_max]
        lista[pos_max] = temp
        print('depois ', lista)
```

```python
alist = [54,26,93,17,77,31,44,55,20]
selection_sort_maior(alist)
print(alist)
```

```python
def selection_sort_menor(lista):
    n = len(lista)
    for slot in range(0, n-1):
        pos_min = slot
        for loc in range(slot+1, n):
            print('slot analisado ', slot, 'item no slot ', lista[slot], 'loc de comparacao ', loc, 'item de comp ', lista[loc])
            if lista[loc] < lista[pos_min]:
                print(lista[loc], ' é menor que ', lista[pos_min])
                pos_min = loc
            else:
                print(lista[loc], ' não é menor que ', lista[pos_min])
        print('menor item de lá pra cá:', lista[pos_min])
        print('antes ', lista)
        temp = lista[slot]
        lista[slot] = lista[pos_min]
        lista[pos_min] = temp
        print('depois ', lista)
```

```python
alist = [54,26,93,17,77,31,44,55,20]
selection_sort_menor(alist)
print(alist)
```
